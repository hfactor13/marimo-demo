```python
from IPython.display import display
import pandas as pd
import ipywidgets as widgets
```

### Example 1

```python
df = pd.DataFrame({
    "name": ["Alice", "Bob", "Charlie"],
    "salary": [70000, 50000, 60000]
})

print("Data loaded")
```

```python
df["salary_k"] = df["salary"] / 1000
print("New column created")
```

```python
avg_salary = df["salary_k"].mean()
print("Average salary (k):", avg_salary)
```

## Example 3

```python
# 1. Create a widget instance
slider = widgets.IntSlider(
    value=5,          # Initial value
    min=0,
    max=10,
    step=1,
    description='Value:', # Label
    disabled=False,
    continuous_update=False, # Update only on release
    orientation='horizontal',
    readout=True,
    readout_format='d'
)

# 2. Define a function to be called when the value changes
def on_value_change(new_value):
    print(f"The new value is: {new_value}")

# 3. Link the function to the 'value' property of the widget
slider.observe(on_value_change, names='value')

# 4. Display the widget
display(slider)
```

```python
on_value_change(slider.value)
```
